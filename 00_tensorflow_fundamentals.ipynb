What we've created so far:

* Scalar: a single number
* Vector: a number with direction (e.g. wind speed and direction)
* Matrix: a 2-dimensional array of numbers
* Tensor: an n-dimensional array of numbers (when n can be any number, a 0-dimensional tensor is a scalar, a 1-dimensional tensor is a vector)

### Creating random tensors

Random tensors are tensors of some abitrary size which contain random numbers

```python
# Create two random (but the same) tensors

random_1 = tf.random.Generator.from_seed(42) # set seed for reproducibility
random_1 = random_1.normal(shape=(3,2))
random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape=(3,2))

# Are they equal?
random_1, random_2, random_1==random_2
```

### Shuffle the order of elements in a tensor

```python
# Shuffle a tensor ( valuable for when you want to shuffle your data so the inherent order does not affect learning)

not_shuffled = tf.constant([[10,7],
                            [3,4],
                            [2,5]])
# Shuffle or non-shuffled tensor
tf.random.shuffle(not_shuffled)
```

**Exercise:** Read through TensorFlow documentation on random seed generation: https://www.tensorflow.org/api_docs/python/tf/random/set_seed and practice writing 5 random tensors and shuffle them.

It looks like if we want our shuffled tensors to be in the same order, we've got to use the global level random seed as well as the operation level random seed:

> Rule 4: If both the global and the operation seed are set: Both seeds are used in conjunction to determine the random sequence.

```python
# Shuffle or non-shuffled tensor
tf.random.set_seed(42) # global level random seed
tf.random.shuffle(not_shuffled, seed=42) # operation level random seed
```

```python
not_shuffled
```

### Getting information from tensors

When dealing with tensors you probably want to be aware of the following attributes:
* Shape
* Rank
* Axis or dimensions
* Size

```python
# Create a rank 4 tensor (4 dimensions)
rank_4_tensor = tf.zeros(shape=[2,3,4,5])
rank_4_tensor
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
2*3*4*5
```

```python
# Get various attributes of our tensor

print("Datatype of every elements:", rank_4_tensor.dtype)
print("Number of dimensions (rank):", rank_4_tensor.ndim)
print("Shape of tensor:", rank_4_tensor.shape)
print("Elements along the 0 axis:", rank_4_tensor.shape[0])
print("Elements along the last axis:", rank_4_tensor.shape[-1])
print("Total number of elements in our tensor:", tf.size(rank_4_tensor))
print("Total number of elements in our tensor:", tf.size(rank_4_tensor).numpy())
```

**Matrix multiplication**

In machine learning, matirx multiplication is one of the most common tensor operations

There are two rules our tensors (or matrices) need to fulfil if we're going to matix multiply them:

1. The inner dimensions must match
2. The resulting matrix has the shape of the inner dimensions

```python
# Matrix multiplication in tensorflow
print(tensor)
tf.matmul(tensor, tensor)
```

```python
tensor * tensor
```

```python
# Matrix multiplication with Python operator "@"
tensor @ tensor
```

```python
# Create a tensor (3,2)
X = tf.constant([[1,2],
                 [3,4],
                 [5,6]])
# Create another (3,2) tensor
Y = tf.constant([[7,8],
                 [9,10],
                 [11,12]])
X, Y
```

```python
# Try to matrix multiply tensors of same shape
X @ Y
```

```python
# Let's change the shape of Y
tf.reshape(Y, shape=(2,3))
```

```python
X.shape, tf. reshape(Y, shape=(2,3)).shape
```

```python
# Try to matrix multiply X by reshaped Y
X @ tf.reshape(Y, shape=(2,3))
```

```python
tf.matmul(X, tf.reshape(Y, shape=(2,3)))
```

```python
# Try change the shape of X instead of Y
tf.matmul(tf.reshape(X, shape=(2,3)),Y)
```

```python
# Can do the same with transpose
X, tf.transpose(X), tf.reshape(X, shape=(2,3))
```

```python
# Try matrix multiplication with transpose rather than reshape
tf.matmul(tf.transpose(X),Y)
```

**The dot product**

Matrix multiplication is also referrred to as the dot product.

You can perform matrix multiplication using:
1. `tf.matmul()`
2. `tf.tensordot()`
3. `@`

```python
X, Y
```

```python
# Perform the dot product on X and Y (requires X or Y to be transposed)

tf.tensordot(tf.transpose(X), Y, axes=1)
```

```python
# Perform matrix multiplicatoin between X and Y (transposed)
tf.matmul(X, tf.transpose(Y))
```

```python
# Perform matrix multiplicatoin between X and Y (reshape)
tf.matmul(X, tf.reshape(Y, shape=(2,3)))
```

```python
# Check the values of Y, reshape Y and tranposed Y
print("Normal Y:")
print(Y, "\n")

print("Y reshaped to (2,3):")
print(tf.reshape(Y, (2,3)), "\n")

print("Y transposed:")
print(tf.transpose(Y))
```

```python
tf.matmul(X, tf.transpose(Y))
```

Generally, when perfomring matrix multiplication on two tnesors and one of the axes doesn't line up, you will transpose (rather than reshape) one of the tensors to satisfy the matrix multiplication rules.

### Aggregating tensors

Aggregating tensors = condensing them from multiple values down to a smaller amount of values.

```python
# Get the abolute values
D = tf.constant([-7, -10])
tf.abs(D)
```

Let's go through the following forms of aggregation:
* Get the minimum
* Get the maximumGet the mean of a tensor
* Get the sum of a tensor

```python
# Create a random tensor with values between 0 and 100 of size 50
E = tf.constant(np.random.randint(0, 100, size=50))
E
```

```python
tf.size(E), E.shape, E.ndim
```

```python
# Find the minimum
tf.reduce_min(E)
```

```python
# Find the maximum
tf.reduce_max(E)
```

```python
# Find the mean
tf.reduce_mean(E)
```

```python
# Find the sum
tf.reduce_sum(E)
```

```python
# Find the variance
# To find the variance of our tensor we need access to tensorflow_probability
import tensorflow_probability as tfp
tfp.stats.variance(E)
```

```python
# Find variance with tf.math
tf.math.reduce_variance(tf.cast(E, dtype=tf.float32))
```
